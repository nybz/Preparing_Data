```python
import numpy as np 
from math import sqrt
```

```python
# function to calculate Cohen's d for independent samples
def cohend(d1, d2):
    # calculate the size of samples
    n1, n2 = len(d1), len(d2)
    # calculate the variance of the samples
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)
    # calculate the pooled standard deviation
    s = sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
    # calculate the means of the samples
    u1, u2 = np.mean(d1), np.mean(d2)
    # calculate the effect size
    return (u1 - u2) / s
```

```python
from pprint import pprint
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV
import os
```

```python
vectorizer = CountVectorizer(# analyzer = 'word',
                             lowercase   = True,
                             ngram_range = (1,3),
                             max_df      = .99,
                             min_df      = .01,
                             #token_pattern = r"(?u)\b\w+\b",
                             # token_pattern = "[a-zA-Z0-9]{3,}",
                             stop_words   = 'english',
                                            # ['000'],
                             max_features = None)
```

```python
review_word_counts = vectorizer.fit_transform(table['Questions'])
```

```python
## check the sparsicity 
    ## Nonzero values_materialize the sparse data 
dense_table = review_word_counts.todense()
# compute sparsicity 
    ## percentage of non-zero cells 
print("Sparsicity: ", ((dense_table > 0).sum()/dense_table.size)*100, "%")
```

```python
# Define Search Param
search_params = {'n_components': [2, 5, 8, 10], 'learning_decay': [.5, .7, .9]}

# Init the Model
lda = LatentDirichletAllocation()

# Init Grid Search Class
model = GridSearchCV(lda, param_grid=search_params)

# Do the Grid Search
model.fit(review_word_counts)
```

```python
# Best Model
best_lda_model = model.best_estimator_

# Model Parameters
print("Best Model's Params: ", model.best_params_)

# Log Likelihood Score
print("Best Log Likelihood Score: ", model.best_score_)

# Perplexity
print("Model Perplexity: ", best_lda_model.perplexity(review_word_counts))
```

```python
lda_model = LatentDirichletAllocation(learning_decay = 0.9,
                                      evaluate_every = -1,
                                      n_components   = 2,
                                      n_jobs = -1)
```

```python
lda_model.fit(review_word_counts)
```

```python
# Log Likelyhood: Higher the better
print("Log Likelihood: ", lda_model.score(review_word_counts))

# Perplexity: Lower the better. Perplexity = exp(-1. * log-likelihood per word)
print("Perplexity: ", lda_model.perplexity(review_word_counts))
pprint(lda_model.get_params())
```

```python
pyLDAvis.enable_notebook()
panel = pyLDAvis.sklearn.prepare(lda_model, review_word_counts, vectorizer, mds='tsne')
panel
```
