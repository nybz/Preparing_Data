## This works for combining all the files to make corpus

```python
import glob
import io
file_list = glob.glob(os.path.join(os.getcwd(), 'C:/Users/2918261/Dropbox/Corner Office Interviews/ComparisonCorpusText/All_Combined', '*.txt')) 

corpus = []

for file_path in file_list:
    with open(file_path) as f_input:
        corpus.append(f_input.read())
        
# print(corpus)
```
